```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * x**3)))
```

```python
def geglu(x, W, W_g):
    gelu_transformed = gelu(W*x)
    gate = sigmoid(W_g*x)
    return gelu_transformed * gate
```

```python
x = np.linspace(-3, 3, 100)
W, W_g = 1.5, 1.5  # Example weights

geglu_output = geglu(x, W, W_g)

# Plot results
plt.figure(figsize=(8, 5))
plt.plot(x, geglu_output, label="GeGLU (Gated GELU)", linewidth=2)
plt.legend()
plt.title("Gated GeLU (GeGLU)")
plt.xlabel("x")
plt.ylabel("Activation")
plt.grid(True)
plt.show()
```
